# Wong Wang Model

```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
# Dynamical Equation Definitions
    
def Hi(xi, a=270, b=108, d=0.154):
    return (a*xi - b)/(1 - np.exp(-d*(a*xi - b)))
```

```python
# default equation params

def default_params(**kwargs):
    p = {}
    
    p['gamma'] = 0.641    
    p['tS'] = 100              # [ms] Time scale of synaptic gating
    p['gE'] = 0.2609           # j11; j22 [nA]
    p['gI'] = -0.0497           # cross inhibition[nA]
    p['Jaext'] = 5.2e-4        # [nA/Hz]
    p['I0'] = 0.3255           # [nA]
    p['mu0'] = 30              # [Hz]
    p['totalTime'] = 2         # [s]
    p['dt'] = .001             # [s]
    p['sigma'] = 0.02
    p['tau0'] = 0.002
    p['coh'] = 0.
    
    p.update(kwargs)

    return p

def simulate(params):
    p = params
    
    t_range = np.arange(0, p['totalTime'], p['dt'])
    n_steps = t_range.size

    # initialize
    s1 = 0.1*np.ones(n_steps)
    s2 = 0.1*np.ones(n_steps)

    x1 = np.ones(n_steps)
    x2 = np.ones(n_steps)
    
    h1 = np.zeros(n_steps)
    h2 = np.zeros(n_steps)
    
    Ieta1 = np.zeros(n_steps)
    Ieta2 = np.zeros(n_steps)
    
    # forward euler
    for k in range(n_steps - 1):

        # calculate I
        Isyn1 = p['gE']*s1[k]  + p['gI']*s2[k] + p['I0'] + p['Jaext']*p['mu0']*(1 + p['coh']/100) + Ieta1[k]
        Isyn2 = p['gE']*s2[k]  + p['gI']*s1[k] + p['I0'] + p['Jaext']*p['mu0']*(1 - p['coh']/100) + Ieta2[k]

        # calculate h (rate)
        h1[k] = Hi(Isyn1)
        h2[k] = Hi(Isyn2)

        # update s
        s1[k+1] = s1[k] + p['dt']*(-s1[k]/p['tS'] + (1 - s1[k])*p['gamma']*h1[k])
        s2[k+1] = s2[k] + p['dt']*(-s2[k]/p['tS'] + (1 - s2[k])*p['gamma']*h2[k])
        
        Ieta1[k+1] = Ieta1[k] + (p['dt']/p['tau0'])*(p['I0']-Ieta1[k]) + np.sqrt(p['dt']/p['tau0'])*p['sigma']*np.random.randn()
        Ieta2[k+1] = Ieta2[k] + (p['dt']/p['tau0'])*(p['I0']-Ieta2[k]) + np.sqrt(p['dt']/p['tau0'])*p['sigma']*np.random.randn()
        
    return h1, h2
```

```python
p = default_params(coh=51)
r1, r2 = simulate(p)

xrange = np.arange(0, p['totalTime'], p['dt'])
fig, ax = plt.subplots(figsize = (4,3),dpi = 150)
ax.plot(xrange, r1, alpha = 0.2, label = 'r1')
ax.plot(xrange, r2, alpha = 0.2, label = 'r2')
ax.plot(xrange, np.ones(xrange.size)*15, '--', label = 'threshold')
ax.set_ylabel('Firing Rate [Hz]')
ax.set_xlabel('Time [s]')
ax.legend()
```
